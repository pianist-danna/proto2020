## やりたいこと
波形描画時の横軸を時間軸にする

```python
import os
import wave
import numpy as np
import matplotlib.pyplot as plt


x = wave.open("C:\\Users\\amane\\OneDrive\\python\\proto2020\\bin\\data\\waves\\OK-1.wav").readframes(-1)
x = np.frombuffer(x,dtype = "int16")
x = x/float((np.power(2,16)/2)-1)
x = x.astype(np.float16)
print(x.dtype)
```

```python
time = np.linspace(0,len(x)/44100,len(x))
plt.plot(time,x)
plt.ylim(-1,1)
plt.show()
```

## わかったこと
- 時間軸を`np.array`で与えればよい
- `np.linespace()`で第二引数`stop`に最後の値、第三引数`num`に要素数を指定する  
    波形データをxとすると、第二引数は`len(x)/sampling_rate`,第三引数は`len(x)`で求まる  
    (要素数は波形データと同じである必要がある)
